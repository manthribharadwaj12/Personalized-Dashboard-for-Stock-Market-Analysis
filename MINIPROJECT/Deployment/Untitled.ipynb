```python
s="                        $bc@gmail.com$pqr@yahoo.co.in$lmn@gmail.com$xyz@gitam.edu$efg@gmail.com$ijk@gitam.edu$ghi@gmail.com$      "
l1=[]
k=[]
for i in range(len(s)):
    if s[i]=='@':
        str=""
        for j in range(i+1,len(s)):
            if s[j]=="$":
                break
            else:
                str+=s[j]
        if str not in l1:
            l1.append(str)
for i in l1:
    k.append(s.count(i))
print("MOST USED DOMAIN IS "+l1[k.index(max(k))])
```
